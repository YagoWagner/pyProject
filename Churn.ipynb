```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

df = pd.read_excel(r'DtScience\Telecom Churn Rate Dataset.xlsx')
display(df)
```

```python
df.columns
```

```python
df.columns = ['ID do cliente', 'Gênero', 'Cidadão Idoso', 'Casado', 'Dependentes',
              'permanência', 'Serviço telefônico', 'Múltiplas linhas', 'Serviço de Internet', 'Segurança online', 
              'Backup online', 'Proteção de dispositivos', 'Suporte técnico',
              'Streaming TV', 'Streaming Movies', 'Contrato', 'Faturamento',
              'Método de pagamento', 'Cobranças mensais', 'Total de cobranças', 'num Tickets de administrador',
              'numTechTickets', 'Churn']
```

```python
df_new = df.drop("ID do cliente", axis=1,inplace=True)
```

```python
display(df.info())
```

```python
# Como estão os nossos cancelamentos(Churn)?
print(df["Churn"].value_counts())
print(df["Churn"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
import plotly.express as px

for coluna in df.columns:
    grafico = px.histogram(df, x=coluna, color="Churn", color_discrete_sequence=["blue", "red"], text_auto=True)
    grafico.show()
```
