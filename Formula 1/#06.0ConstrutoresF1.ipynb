## Sobre o conjunto de dados
### Contexto
A Fórmula 1 (também conhecida como Fórmula 1 ou F1981) é a mais alta classe de automobilismo internacional de monopostos sancionada pela Fédération Internationale de l'Automobile (FIA) e de propriedade da Liberty Media, uma empresa americana de mídia de massa, através de sua subsidiária integral, o Formula One Group. O Campeonato Mundial de Pilotos, que se tornou o Campeonato Mundial de Fórmula Um da FIA em 1950, tem sido uma das principais formas de corrida em todo o mundo desde sua temporada inaugural em <>. A palavra "fórmula" no nome refere-se ao conjunto de regras às quais os carros de todos os participantes devem estar em conformidade. Uma temporada de Fórmula Um consiste em uma série de corridas, conhecidas como Grandes Prêmios (francês para "grandes prêmios" ou "grandes prêmios"), que ocorrem em todo o mundo em circuitos construídos especificamente e em vias públicas.

### Conteúdo
Este conjunto de dados contém os dados de todos os grandes prêmios de Fórmula 1950 realizados desde <>. Também inclui os resultados das corridas, bem como a classificação dos pilotos de cada temporada e os construtores em pé.

### Inspiração
A coleta e análise de dados é uma das partes mais vitais das corridas de Fórmula 1. Com base nos dados coletados por cada equipe, eles desenvolvem seu próprio carro, para inovar a tecnologia que usam. Sem a análise de dados em tempo real, a F1 que conhecemos hoje poderia ser muito diferente. Centenas de pontos de dados por segundo são gerados em todas as áreas-chave dos carros. Enquanto isso, engenheiros e analistas de dados, tanto no circuito quanto nas fábricas (que estão a milhares de quilômetros de distância), competem para extrair o máximo de informações possível e armar essas informações para lutar pela vitória.
