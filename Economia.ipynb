Os Estados Unidos são uma economia mista altamente desenvolvida. É a maior economia do mundo por PIB nominal e a segunda maior por paridade de poder de compra (PPC), atrás da China. Tem o sétimo maior PIB per capita (nominal) do mundo e o oitavo maior PIB per capita (PPC) a partir de 2022. Os EUA representaram 24,7% da economia global em 2022 em termos nominais e cerca de 15,5% em termos de PPC. O dólar dos EUA é a moeda de registro mais usada em transações internacionais e é a principal moeda de reserva do mundo, apoiada por um grande mercado de títulos do Tesouro dos EUA, seu papel como padrão de referência para o sistema de petrodólares e seu dólar euro vinculado. Vários países usam-na como moeda oficial e noutros é a moeda de facto.

```python
def Multi_boxplot(dataframe: pd.DataFrame) -> plt.show:
    """
     Função que gera vários boxplots a partir de um dataframe dado.

     Parâmetros
     ----------
     dataframe: pd.DataFrame
         O dataframe que contém os dados para gerar os boxplots.

     devoluções
     -------
     plt.show
         Mostra os boxplots gerados.

     """
    plt.figure(figsize=[15,18])
    n = 1
    for f in dataframe.columns:
        mediana_agrupada = dataframe[f].median()
        maximo = dataframe[f].max()
        plt.subplot(5,2,n)
        sns.boxplot(x = dataframe[f], data = dataframe, notch=True)
        plt.axvline(x = mediana_agrupada, color='r', linestyle='-', label=f'Mediana {round(mediana_agrupada, 4)}')
        plt.axvline(x = maximo, color='g', linestyle='-', label=f'Maximo {round(maximo, 4)}')
        plt.legend()
        plt.title(f'Boxplot {f} ')
        n = n+1
    plt.tight_layout()
    return plt.show()
```

```python
Multi_boxplot(dfnew)
```

```python
def Multi_lineGraph(dataframe: pd.DataFrame, a: str,
                    b: str, c: str) -> plt.show:
    """
     Crie um gráfico de linhas com os ejetos verticais para visualizar as variáveis na função de uma terceira variável.

     Parâmetros:
     -----------
     dataframe: pd.DataFrame
         O DataFrame que contém os dados para visualização.
     a: str
         O nome da variável que será mostrado no eje e direito do gráfico.
     b: str
         O nome da variável que será mostrado no eje e será excluído do gráfico.
     c: str
         O nome da variável que se mostrará no eje x do gráfico.

     Retorna:
     --------
     Nenhum
         A função mostra o gráfico, mas não deve ter nada.

     Exemplos:
     ----------
     >>> Multi_lineGraph(df, 'Variável1', 'Variável2', 'Variável3')
     """

    fig, axes = plt.subplots(figsize=(10,6))
    axes.plot(dataframe[c], dataframe[b], color='b', marker='o')
    axes.set_xlabel(f'{c}')
    axes.set_ylabel(f'{b}', color='b')
    twin_axes = axes.twinx()
    twin_axes.plot(dataframe[c], dataframe[a], color='r', marker='*')
    twin_axes.set_ylabel(f'{a}', color='r')
    return plt.show()
```
