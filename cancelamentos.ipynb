```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

tabela = pd.read_excel(r'DtScience\Telecom Churn Rate Dataset.xlsx')
display(tabela)
```

```python
tabela1 = tabela.drop("customerID", axis=1,inplace=True)
```

```python
display(tabela.info())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for coluna in tabela.columns:
    tabela[coluna] = le.fit_transform(tabela[coluna].astype(str))


print(tabela.info())
```

```python
print(tabela["Churn"].value_counts())
print(tabela["Churn"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
display(tabela.groupby("Contract").mean(numeric_only=True))
```

```python
import plotly.express as px

for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="Churn", text_auto=True)
    grafico.show()
```

```python
tabela = tabela[tabela["Contract"] != 0]

display(tabela["Churn"].value_counts())
display(tabela["Churn"].value_counts(normalize=True).map("{:.1%}".format))
```
