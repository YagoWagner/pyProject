```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

filmes = pd.read_csv(r'DtScience\IMDB Top 250 Movies.csv')
filmes.head(3)
```

```python
filmes.columns = ['classificação', 'nome', 'ano', 'avaliação', 'gênero', 'certificado', 'tempo_filme',
        'slogan', 'orçamento', 'bilheteria', 'elenco', 'diretores', 'escritores']
```

Para i na coluna 'Gênero'

real_gen: vai ser todos os gêneros em uma váriavel

all_gen: vai pegar o real_gen e separar todos os itens com uma vírgula com cada um deles, puxando do index 1 até o final

unique_genres: lista todos os itens em um set com o all_gen, onde todos os itens não vão ser repetidos

```python
real_gen = ""
for i in filmes['gênero']:                
    real_gen = real_gen+","+i            
all_gen = real_gen.split(',')[1:]        
unique_genres = list(set(all_gen))    

print(i)
print(real_gen)
print(all_gen)
print(unique_genres)
```

'data' é uma lista

O for loop pega cada item (i) de unique_genres e começa outro loop com um contador igual a zero.

    O contador (c) é para contarmos quantos gêneros de filmes separados nós temos. 

O segundo for loop, se o item 'i' estiver em 'j' que é um for loop em filme['gênero'] o contador conta 1 em um loop com todos os 250 itens

Todos esses dados serão juntadas em várias listas dentro de uma lista maior separadas por gênero separado (i) e quantidade (c)

```python
data = []
for i in unique_genres:
    c = 0
    for j in filmes['gênero']:
        if i in j:
            c+=1
    data.append([i, c])

print(c)
display(data)
```

pd.DataFrame() vai criar uma nova tabela açossiada a variável "gen_df", onde pegaremos a variável 'data' que usaremos para criar a nova tabela, precisaremos colocar 2 novas colunas: 'gênero', 'frequência', que são os 2 valores distintos se tivesse 3 valores precisariamos nomear 3 colunas.

O sort_values queremos que a tabela mostre a partir da coluna frequência em ordem decrescente.

    by: Único/Lista de nomes de coluna para classificar o Quadro de Dados
    ascending: valor booleano que classifica o quadro de dados em ordem crescente se True e decrescente se False. 
    inplace: valor booleano. Faz as alterações no próprio quadro de dados passados se True, deve modificar o DataFrame em vez de criar um novo.

gen_df.head(5): vai pegar os 5 primeiros elementos da nova tabela.

```python
gen_df = pd.DataFrame(data, columns=['gênero', 'frequência'])
gen_df.sort_values(by = 'frequência', ascending = False, inplace = True)
gen_df = gen_df.head(5)
display(gen_df)
```

```python
plt.figure(figsize=(10,5),dpi = 100)
sns.barplot(data = gen_df, x = 'gênero', y = 'frequência', palette='Paired')
plt.title('Top 5 gêneros mais populares')
plt.grid(True)
plt.show()
```

filmes.groupby('certificado').size(): Agrupa toda a coluna 'certificado' por tamanho dos elementos '.size()'

.sort_values(ascending=False).head(5): Vai pegar os resultados em ordem decrescente e pegar os 5 primeiros

.plot(kind = 'bar', color = sns.color_palette('Set2')): Cria um gráfico de barras com uma paleta de cores especifica.

```python
plt.figure(figsize=(10,5))
filmes.groupby('certificado').size().sort_values(ascending=False).head(5).plot.bar(color = sns.color_palette('Set2'))
plt.xlabel('Certificado')
plt.ylabel('Frequência')
plt.title("Os 5 melhores certificados")
plt.grid(True)
plt.show()

print(filmes.groupby('certificado').size().sort_values(ascending=False))
```

```python
plt.figure(figsize=(16,6))
filmes.groupby('diretores').size().sort_values(ascending=False).head(10).plot.bar(color = sns.color_palette('Set3'))
plt.xlabel("Nome do Diretor")
plt.ylabel("Número de filmes")
plt.title("Top 10 diretores que dirigiram mais filmes no IMDB top 250 filmes")
plt.grid(True)
plt.show()

print(filmes.groupby('diretores').size().sort_values(ascending=False).head(10))
```

```python
filmes['ano'].sort_values()
# O ano é organizado em ordem
# ano varia de 1921 a 2022
# vamos fazer grupo como 1921 - 1940, 1941- 1960, 1961 - 1980, 1981 - 2000, 2001 - 2022
```

```python
_21to40 = 0
_41to60 = 0
_61to80 = 0
_81to00 = 0
_01to22 = 0

for i in filmes['ano']:
    if i>1920 and i<=1940:
        _21to40+=1
    elif i>1940 and i<=1960:
        _41to60+=1
    elif i>1960 and i<=1980:
        _61to80+=1
    elif i>1980 and i<=2000:
        _81to00+=1
    else: 
        _01to22+=1
```

```python
data = [_21to40, _41to60, _61to80, _81to00, _01to22]
label = ['1921 - 1940', '1941- 1960', '1961 - 1980', '1981 - 2000', '2001 - 2022']
plt.figure(figsize=(12,6))
plt.pie(data,labels=label,colors=sns.color_palette('Paired'),autopct='%.2f%%')
plt.title("Porcentagem de filmes com base no ano")
plt.show()
```
