```python
import pandas as pd


df = pd.read_csv(r'..\DtScience\statsfinal.csv')
df = df.drop("Unnamed: 0", axis=1) # Exclui a COLUNA "Unnamed: 0"
colunas_p_drop = ['S-P1', 'S-P2', 'S-P3', 'S-P4']
df = df.drop(colunas_p_drop, axis=1)
display(df)
```

```python
df['Day'] = df['Date'].apply(lambda x: x.split('-')[0])
df['Month'] = df['Date'].apply(lambda x: x.split('-')[1])
df['Year'] = df['Date'].apply(lambda x: x.split('-')[2])
df = df.drop('Date', axis=1)
df
```

```python
data = df.query("Year != '2010' and Year != '2023'")
data
```

```python
def mes_31_dez(df, months):
    m31_data = df[df['Month'].isin(months)]
    return m31_data

dezembro = mes_31_dez(data, ['12'])
dezembro
```

```python
estimativa_vendas_ano = dezembro.groupby('Year')[['Q-P1', 'Q-P2', 'Q-P3', 'Q-P4']].sum()
estimativa_vendas_ano
```

```python
media_vendas_dez_ano = dezembro.groupby('Year')[['Q-P1', 'Q-P2', 'Q-P3', 'Q-P4']].mean()
media_vendas_dez_ano
```

```python
estimativa_dezembro = dezembro.groupby('Day')[['Q-P1', 'Q-P2', 'Q-P3', 'Q-P4']].sum()
estimativa_dezembro
```

```python
media_dezembro = dezembro.groupby('Day')[['Q-P1', 'Q-P2', 'Q-P3', 'Q-P4']].mean()
media_dezembro
```
