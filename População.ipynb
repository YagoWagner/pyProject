## Sobre o conjunto de dados

### Sobre o conjunto de dados

Estes são dados sobre a população da Terra coletados de 4 locais que fornecem estatísticas sobre a população da Terra.

Assim, os valores para os prefixos contam e taxa de crescimento:

zw - zhujiworld.com;

wm - worldometers.info;

mt - macrotrends.net;

cm - countrymeters.info.

### Descrições das colunas principais

ano - ano;

contagem - tamanho da população;

taxa de crescimento - taxa de crescimento.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import bar_chart_race as bcr
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r'DtScience\finaldata.csv')
display(df)
```

```python
fig, ax1 = plt.subplots(figsize=(23,13))
fig, ax2 = plt.subplots(figsize=(23,13))

sns.lineplot(data=df, x='year', y='count_zw', marker="o", ax=ax1, label='zw',alpha=0.4, linewidth=1)
sns.lineplot(data=df, x='year', y='count_wm', marker="o", ax=ax1, label='wm',alpha=0.4, linewidth=1)
sns.lineplot(data=df, x='year', y='count_mt', marker="o", ax=ax1, label='mt',alpha=0.4, linewidth=1)
sns.lineplot(data=df, x='year', y='count_cm', marker="o", ax=ax1, label='cm',alpha=0.4, linewidth=1).set(title="Indicadores diferentes no mesmo gráfico (população)")

sns.lineplot(data=df, x='year', y='growthrate_zw', marker="o", ax=ax2, label='zw',alpha=0.4)
sns.lineplot(data=df, x='year', y='growthrate_wm', marker="o", ax=ax2, label='wm',alpha=0.4)
sns.lineplot(data=df, x='year', y='growthrate_mt', marker="o", ax=ax2, label='mt',alpha=0.4)
sns.lineplot(data=df, x='year', y='growthrate_cm', marker="o", ax=ax2, label='cm',alpha=0.4).set(title="Indicadores diferentes no mesmo gráfico (taxa de crescimento)")

plt.show()
```
