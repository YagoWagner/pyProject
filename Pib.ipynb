## Sobre o conjunto de dados

Este conjunto de dados contém os dados históricos de 1960 a 2021 do PIB por país, além disso, sua taxa de crescimento por ano é calculada.

Os dados são obtidos a partir dos dados do Banco Mundial, o conjunto de dados é baixado, um pré-processamento foi realizado no qual dados geográficos como regiões, sub-regiões foram adicionados e a variação percentual por ano e país foi calculada.

O principal objetivo deste conjunto de dados é servir como fonte de dados para a análise populacional que estou desenvolvendo para entender melhor os fatores que afetam o crescimento populacional.

```python
import pandas as pd
import numpy as np
import bar_chart_race as bcr

import warnings
warnings.filterwarnings("ignore")

data = pd.read_csv(r'DtScience\countries_gdp_hist.csv', sep=';')
display(data)
```

```python
pivot_df = data2.pivot(index='year', columns='country_name', values='total_gdp_million')
display(pivot_df)
```
