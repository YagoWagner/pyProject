# Conjunto de dados de preço do carro
O conjunto de dados Carros contém 211 dos 24 recursos dos carros.

## Sobre o conjunto de dados
O conjunto de dados Carros contém 211 de24 recursos de carros. onde os recursos foram divididos aproximadamente em uma divisão de 50-50. Os recursos são tipicamente no nível de Marca, Modelo, Ano, por exemplo, 2012 Tesla Model S ou 2012 BMW M3 cupê.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

carro = pd.read_csv(r'DtScience\car_price.csv')
display(carro)
```

```python
carro.info()
```

```python
carro.columns = ['Nome do carro', 'carcaça', 'roda', 'localização do motor', 'tipo de combustível',
        'aspiração', 'número da porta', 'número do cilindro', 'tipo de motor','sistema de combustível', 'simbolização',
        'distância entre eixos', 'comprimento do carro', 'largura do carro', 'altura do carro', 'peso do meio-fio', 
        'tamanho do motor', 'taxa de furo', 'stroker', 'taxa de compressão', 'cavalos de força', 
        'rpm de pico', 'mpg na cidade', 'mpg na rodovia', 'preço']
```

```python
carro.info()
```

```python
preco = carro[["Nome do carro","preço"]].sort_values(by="preço", ascending=False)[:10]
preco
```

```python
preco[:10].plot.barh(x='Nome do carro', y='preço',color='green',title='TOP 10 Carros Mais Caros');
```

```python
print(carro['roda'].value_counts(normalize= True).map("{:.1%}".format))
```

```python
montadora = ''
for i in carro['Nome do carro']:
    montadora = montadora+','+ i
all_car = montadora.split(',')[1:]
veiculo = list(set(all_car))

print(montadora)
print(all_car)
print(veiculo)
```

```python
data = []
for i in veiculo:
    c = 0
    for j in carro['Nome do carro']:
        if i in j:
            c+=1
    data.append([i, c])

display(data)
```

```python
dfcar = pd.DataFrame(data, columns=['Veiculo', 'Quantidade'])
dfcar.sort_values(by = 'Quantidade', ascending = False, inplace = True)
dfcar = dfcar.head(10)
display(dfcar)
```

```python
carro['Montadora'] = carro['Nome do carro'].apply(lambda x: x.split(' ')[0])
```

```python
carro.Montadora.replace(to_replace=['vw', 'vokswagen', 'Nissan', 'toyouta', 'maxda', 'porcshce'], 
value=['volkswagen', 'volkswagen', 'nissan', 'toyota', 'mazda', 'porsche'], inplace=True)
```

```python
vendas_total = carro.groupby('Montadora')['preço'].sum().sort_values(ascending=False).reset_index().head(10)
display(vendas_total)
```

```python
plt.scatter(vendas_total["Montadora"],vendas_total["preço"])
x0 = vendas_total["Montadora"][0]
y0 = vendas_total["preço"][0]
x1 = vendas_total["Montadora"][9]
y1 = vendas_total["preço"][9]
plt.plot([x0,x1],[y0,y1],"r")
plt.show()
```

```python
plt.figure(figsize=(16,6))
carro.groupby('Montadora').size().sort_values(ascending=False).head(10).plot.barh(color = sns.color_palette('Set3'))
plt.xlabel("Quantidade")
plt.ylabel("Nome do carro")
plt.title("Top 10 Montadoras Com Mais Carros")
plt.grid(True)
plt.show()

print(carro.groupby('Montadora').size().sort_values(ascending=False).head(10))
```
